```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("Sample - Superstore.csv",encoding="unicode_escape")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# coverting the data type to datetime

df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
df.info()
```

```python
df["Order Year"] = df["Order Date"].dt.year
df["Order Month"] = df["Order Date"].dt.month 
df["Order Day of Week"] = df["Order Date"].dt.dayofweek 
```

```python
df.head()
```

```python
# Monthly analysis

monthly_trend = df.groupby(["Order Month"],as_index=False)["Sales"].sum().sort_values(["Sales"],ascending=False)
sns.barplot(data=monthly_trend,x="Order Month",y="Sales")
plt.title("Sales_by_month")
plt.show()
```

```python
# Sale by category

sales_by_category = df.groupby(["Category"],as_index=False)["Sales"].sum().sort_values(["Sales"],ascending=False)
sales_by_category
```

```python
sns.barplot(data=sales_by_category,x="Category",y="Sales")
plt.title("Sales_by_category")
plt.show()
```

```python
# Sale by sub-category

sales_by_subcategory = df.groupby(["Sub-Category"],as_index=False)["Sales"].sum().sort_values(["Sales"],ascending=False)
sales_by_subcategory
```

```python
sns.barplot(data=sales_by_subcategory,x="Sub-Category",y="Sales")
sns.set(rc={"figure.figsize":(20,5)})
plt.title("Sales_by_subcategory")
plt.show()
```

```python
# profit analysis by month

profit_by_month =df.groupby(["Order Month"],as_index=False)["Profit"].sum().sort_values(["Profit"],ascending=False)
profit_by_month
```

```python
sns.barplot(data=profit_by_month,x="Order Month",y="Profit")
sns.set(rc={"figure.figsize":(20,5)})
plt.title("Profit_by_month")
plt.show()
```

```python
# Profit by category

profit_by_category = df.groupby(["Category"],as_index=False)["Profit"].sum().sort_values(["Profit"],ascending=False)
profit_by_category
```

```python
sns.barplot(data=profit_by_category,x="Category",y="Profit")
sns.set(rc={"figure.figsize":(20,5)})
plt.title("Profit_by_category")
plt.show()
```

```python
# Profit by sub-category

profit_by_subcategory = df.groupby(["Sub-Category"],as_index=False)["Profit"].sum().sort_values(["Profit"],ascending=False)
profit_by_subcategory
```

```python
sns.barplot(data=profit_by_subcategory,x="Sub-Category",y="Profit")
sns.set(rc={"figure.figsize":(20,5)})
plt.title("Profit_by_subcategory")
plt.show()
```

```python
# Profit by segment

segment = df.groupby(["Segment"],as_index=False)["Profit"].sum().sort_values(["Profit"],ascending=False)
segment
```

```python
sns.barplot(data=segment,x="Segment",y="Profit")
sns.set(rc={"figure.figsize":(20,5)})
plt.title("Segment_by_profit")
plt.show()
```

```python
# Sales by segment

segment = df.groupby(["Segment"],as_index=False)["Sales"].sum().sort_values(["Sales"],ascending=False)
segment
```

```python
sns.barplot(data=segment,x="Segment",y="Sales")
sns.set(rc={"figure.figsize":(20,5)})
plt.title("Segment_by_sales")
plt.show()
```

```python
# Sales to profit ratio

df["Sales_to_Profit_Ratio"] = df["Sales"]/df["Profit"]
```
